## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.linear_model import LogisticRegression
```

## Data Collection And Preprocessing

```python
# Load the dataset
cancer_data = sklearn.datasets.load_breast_cancer()
print(cancer_data)
```

```python
# Loading the data insto dataframe
cancer_dataset = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
cancer_dataset.head()
```

```python
# Adding the target column to the dataframe
cancer_dataset["label"] = cancer_data.target
cancer_dataset.head()
```

```python
# Shape of the dataset
cancer_dataset.shape
```

```python
# Basic information about the dataset
cancer_dataset.info()
```

```python
# Check missing values
cancer_dataset.isna().sum()
```

```python
# Statistical measures
cancer_dataset.describe()
```

```python
# Corelation
cancer_dataset.corr()
```

```python
cancer_dataset.groupby("label").mean()
```

## Features Target Split

```python
x = cancer_dataset.drop(columns=["label"], axis=1)
y = cancer_dataset["label"]
```

## Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

## Train The Model

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

## Model Evaluation

```python
# Training data
y_pred_train = model.predict(x_train)

accuracy = accuracy_score(y_train, y_pred_train)
print("Accuracy on training data: ", accuracy)
```

```python
# Testing data
y_pred_test = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred_test)
print("Accuracy on testing data: ", accuracy)
```

```python
confusion_matrix(y_test, y_pred_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot=True)
plt.show()
```
